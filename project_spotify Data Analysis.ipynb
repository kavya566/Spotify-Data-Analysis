```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from wordcloud import WordCloud, STOPWORDS
```

```python
most_popular=df_tracks.query('popularity>50',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True, vmin= -1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("correlation HeatMap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
top_ten_instrumental_tracks=df_tracks[["instrumentalness","name","artists"]].sort_values(by="instrumentalness",ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data=top_ten_instrumental_tracks,autopct='%1.2f%%',labels=top_ten_instrumental_tracks.name)
plt.title("Top 10 instrumentalness tracks")
plt.show()
```

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax =plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax= ax,errwidth=False).set(title="year vs Duration")
plt.xticks(rotation=90)
```

```python
interest_feature_cols=["tempo","loudness","acousticness","danceability","duration",
                       "energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for feature_col in interest_feature_cols:
  pos_data =df_tracks[df_tracks["mode"]==1][feature_col]
  neg_data =df_tracks[df_tracks["mode"]==0][feature_col]
  plt.figure(figsize=(12,7))
  sns.histplot(pos_data, bins=30, label='positive',color='green')
  sns.histplot(neg_data, bins=30, label='negative',color="red")
  plt.legend(loc="upper right")
  plt.title("Positive and Negative Histogram plot for features_col")
  plt.show()
```
